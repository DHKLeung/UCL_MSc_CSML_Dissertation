```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = [0.99997,
0.99973,
0.99958,
0.99980,
0.99942,
0.99595
]
test = [0.99480,
0.99370,
0.99440,
0.99540,
0.99480,
0.98970
]
standard_train = 0.99917 * np.ones((6,))
standard_test = 0.99320 * np.ones((6,))
xticks = [0.25, 0.5, 0.75, 0.9, 0.99, 1]

plt.figure(figsize=(14, 8))
plt.plot(np.arange(6) + 1, train, 'ro-')
plt.plot(np.arange(6) + 1, standard_train, 'r-.')
plt.plot(np.arange(6) + 1, test, 'bo-')
plt.plot(np.arange(6) + 1, standard_test, 'b-.')

plt.title('MixUp Weighted Augmentation; MNIST', fontsize=17)
plt.legend(['train (weighted augmentation)', 'train (standard)', 'test (weighted augmentation)', 'test (standard)'], fontsize=13)
plt.xticks(np.arange(6) + 1, xticks, fontsize=13)
plt.xlabel('$\gamma$', fontsize=13)
plt.ylabel('Accuracy', fontsize=13)

plt.show()
```
